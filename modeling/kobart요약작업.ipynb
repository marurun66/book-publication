```python
import json
import torch
from transformers import BartForConditionalGeneration, PreTrainedTokenizerFast

# âœ… í•™ìŠµëœ KoBART ëª¨ë¸ ë¡œë“œ
model_checkpoint = "./kobart_best_model"  # ğŸ”¹ Best ëª¨ë¸ì´ ì €ì¥ëœ ì²´í¬í¬ì¸íŠ¸ í´ë”
model = BartForConditionalGeneration.from_pretrained(model_checkpoint)
tokenizer = PreTrainedTokenizerFast.from_pretrained(model_checkpoint)
```

```python
# âœ… ëª¨ë¸ì„ GPUë¡œ ì´ë™ (ê°€ëŠ¥í•˜ë©´)
device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)
```

```python
# âœ… ìš”ì•½ í•¨ìˆ˜ ì •ì˜ (í”„ë¡¬í”„íŠ¸ í¬í•¨)
def generate_summary(description):
    if not description.strip():
        return "ì¤„ê±°ë¦¬ ì—†ìŒ"  # ğŸ”¹ ì„¤ëª…ì´ ë¹„ì–´ ìˆìœ¼ë©´ ì¤„ê±°ë¦¬ ì—†ìŒ ë°˜í™˜
    
    # âœ… í”„ë¡¬í”„íŠ¸ í¬í•¨
    prompt = f"ë‹¤ìŒ ë¬¸ì¥ì„ ìš”ì•½í•˜ì„¸ìš”. ë“±ì¥ì¸ë¬¼ì˜ ì§ì—…, ìŠ¤í† ë¦¬ ì „ê°œ, ì£¼ìš” ì‚¬ê±´ì„ í¬í•¨í•˜ê³ , ì‘ê°€ì˜ ì—°í˜ê³¼ ìˆ˜ìƒì´ë ¥ì€ ì œì™¸í•˜ì„¸ìš”:\n\n{description}"

    # ğŸ”¹ ì…ë ¥ í…ìŠ¤íŠ¸ í† í°í™”
    input_ids = tokenizer(prompt, return_tensors="pt", max_length=1024, truncation=True).input_ids.to(device)

    # ğŸ”¹ ëª¨ë¸ë¡œ ìš”ì•½ ìƒì„±
    with torch.no_grad():
        summary_ids = model.generate(
            input_ids, 
            max_length=300,  # âœ… ìš”ì•½ ê¸¸ì´ ì¡°ì •
            min_length=150,  # âœ… ë„ˆë¬´ ì§§ì€ ìš”ì•½ ë°©ì§€
            num_beams=5,  # âœ… ë¹” ì„œì¹˜ ê°•í™”
            early_stopping=True,  # âœ… ì¡°ê¸° ì¢…ë£Œ (ë°˜ë³µ ë°©ì§€)
            temperature=0.7,  # âœ… ë‹¨ì–´ ì„ íƒ ë‹¤ì–‘ì„± ì¦ê°€
            top_k=50,  # âœ… í™•ë¥  ë†’ì€ 50ê°œ ë‹¨ì–´ ì¤‘ ì„ íƒ
            top_p=0.9,  # âœ… í™•ë¥  ë¶„í¬ì˜ 90%ë¥¼ í¬í•¨í•œ ë‹¨ì–´ ì„ íƒ
            length_penalty=1.5,  # âœ… ê¸´ ë¬¸ì¥ì¼ìˆ˜ë¡ ë” ì ì ˆí•˜ê²Œ ìš”ì•½í•˜ë„ë¡ ìœ ë„
            repetition_penalty=1.5  # âœ… ë°˜ë³µì ì¸ ë‹¨ì–´ ìƒì„±ì„ ì–µì œ
        )
    
    # ğŸ”¹ ìš”ì•½ë¬¸ ë””ì½”ë”©
    summary_text = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
    
    return summary_text

# âœ… ì†Œì„¤ ë°ì´í„° ë¡œë“œ
with open("unlabeled_books_no_summary.json", "r", encoding="utf-8") as f:
    novels_data = json.load(f)
```

```python
# âœ… ëª¨ë“  ì†Œì„¤ì˜ descriptionì„ ìš”ì•½
for novel in novels_data:
    novel["summary"] = generate_summary(novel["description"])

# âœ… ìƒˆë¡œìš´ JSON íŒŒì¼ë¡œ ì €ì¥
output_file = "unlabeled_books_with_summary.json"
with open(output_file, "w", encoding="utf-8") as f:
    json.dump(novels_data, f, ensure_ascii=False, indent=4)

print(f"âœ… ìš”ì•½ ì™„ë£Œ! ê²°ê³¼ê°€ '{output_file}' íŒŒì¼ì— ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤.")
```
